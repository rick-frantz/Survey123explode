```python
# import
from arcgis.gis import GIS
import getpass
import arcgis.features
import pandas as pd

# Connection Variables
Organization = "https://dublinohio.maps.arcgis.com/"
#Organization = input("What's your oranization? ")
User = "DublinOhio"
Password = getpass.getpass('Password: ')

# Connection 
gis = GIS(Organization, User, Password)
gis
```

```python
#Load the winter blues layer

itemId = '51a1f6a4c5b148aa9b1159de0ed4114b'
fs =gis.content.get(itemId)
display(fs)
lyrs = fs.layers
for lyr in lyrs:
    print(lyr.properties.name)
```

```python
df = lyr.query(where='1=1',as_df='true') #querying without any conditions returns all the features
#df = lyr.query(where='1=1')
df
```

```python
#testdf.explode('NutritionPick')
#list1 = list(map(str, testdf['NutritionPick'][0].split(",")))

dataColumns = [
            'objectid',
            'globalid',
            'neighborhood',
            'PhyscalActivityPick',
            'PurposePick',
            'MindfulPick',
            'NutritionPick',
            'NumActivitiesSelected',
            'NumPurposeSelected',
            'NumMindfulSelected',
            'NumNutritionSelected2',
            'Creator',
            'EditDate',
            'Editor',
            'RecordUse']

evaldf = pd.DataFrame(columns = dataColumns)

ActList,PurList,MindList,NutrList = "Default","Default","Default","Default"

for index, row in df.iterrows():
    print("Starting index",index)
    maxSelected = max(df.iloc[index]['NumActivitiesSelected'],df.iloc[index]['NumPurposeSelected'],df.iloc[index]['NumMindfulSelected'],df.iloc[index]['NumNutritionSelected2'])
    if maxSelected > 1:
        #print(index,maxSelected)
        if df.iloc[index]['NumActivitiesSelected'] > 1:
            ActList = list(map(str, df['PhyscalActivityPick'][index].split(",")))
        elif df.iloc[index]['NumActivitiesSelected'] == 1:
            ActList = list(map(str, df['PhyscalActivityPick'][index].split(",")))
        else:
            ActList = ['default']
        if df.iloc[index]['NumPurposeSelected'] > 0:
            PurList = list(map(str, df['PurposePick'][index].split(",")))
        elif df.iloc[index]['NumPurposeSelected'] == 1:
            PurList = df.iloc[index]['PurposePick']
        else:
            PurList = ['default']
        if df.iloc[index]['NumMindfulSelected'] > 0:
            MindList = list(map(str, df['MindfulPick'][index].split(",")))
        elif df.iloc[index]['NumMindfulSelected'] == 1:
            MindList = df.iloc[index]['MindfulPick']
        else:
            MindList = ['default']
        if df.iloc[index]['NumNutritionSelected2'] > 0:
            NutrList = list(map(str, df['NutritionPick'][index].split(",")))
        elif df.iloc[index]['NumNutritionSelected2'] == 1:
            NutrList = df.iloc[index]['NutritionPick']
        else:
            NutrList = ['default']
        #print("Activities selected:",ActList)
        #print("Purpose selected:",PurList)
        #print("Mindful selected:",MindList)
        #print("Nutrition selected:",NutrList)
        for m in range(maxSelected):
            evaldf = evaldf.append({ 
            'objectid':df.iloc[index]['objectid'],
            'globalid':df.iloc[index]['globalid'],
            'neighborhood':df.iloc[index]['neighborhood'],
            'PhyscalActivityPick':ActList[m] if m < len(ActList) else "default",
            'PurposePick':PurList[m] if m < len(PurList) else "default",
            'MindfulPick':MindList[m] if m < len(MindList) else "default",
            'NutritionPick':NutrList[m] if m < len(NutrList) else "default",
            #'NutritionPick':NutrList[m] if m < (0 if not (NutrList) else len(NutrList)) else "default",
            'NumActivitiesSelected':df.iloc[index]['NumActivitiesSelected'],
            'NumPurposeSelected':df.iloc[index]['NumPurposeSelected'],
            'NumMindfulSelected':df.iloc[index]['NumMindfulSelected'],
            'NumNutritionSelected2':df.iloc[index]['NumNutritionSelected2'],
            'Creator':"Pandas:Expanded",
            'EditDate':df.iloc[index]['EditDate'],
            'Editor':"Pandas",
            'RecordUse':"Pandas"
            }, ignore_index=True)
        #testdf['RecordUse'][index] = "Expanded"
        ActList,PurList,MindList,NutrList = "Default","Default","Default","Default"
    else:
        print("This record did not need to be expanded so we will append it as is")
        evaldf = evaldf.append({ 
            'objectid':df.iloc[index]['objectid'],
            'globalid':df.iloc[index]['globalid'],
            'neighborhood':df.iloc[index]['neighborhood'],
            'PhyscalActivityPick':df.iloc[index]['PhyscalActivityPick'],
            'PurposePick':df.iloc[index]['PurposePick'],
            'MindfulPick':df.iloc[index]['MindfulPick'],
            'NutritionPick':df.iloc[index]['NutritionPick'],
            'NumActivitiesSelected':df.iloc[index]['NumActivitiesSelected'],
            'NumPurposeSelected':df.iloc[index]['NumPurposeSelected'],
            'NumMindfulSelected':df.iloc[index]['NumMindfulSelected'],
            'NumNutritionSelected2':df.iloc[index]['NumNutritionSelected2'],
            'Creator':"Original",
            'EditDate':df.iloc[index]['EditDate'],
            'Editor':"Pandas",
            'RecordUse':"Pandas"
            }, ignore_index=True)
    print("Finished record index",index)
    #testdf['RecordUse'][index] = "Original"

evaldf
```
